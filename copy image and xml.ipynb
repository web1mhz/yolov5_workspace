```python
def convert_xml_yolo(f, sub_dir_name):   
    
    f.write(sub_dir_name)    
```

```python
def xml_2_yolo_txt(src_xml_path, dest_xml_path, sub_dir_name):
    
    label_txt= os.path.splitext(dest_xml_path)[0] + '.txt'
    
    with open(label_txt, 'w') as f:        
        convert_xml_yolo(f, sub_dir_name)       
        
    print(label_txt)
```

```python
for dir_name in data_dir:
    file_path = os.path.join(wildlife_dir , dir_name)
    print("파일경로" , file_path)    
    sub_dir = os.listdir(file_path)
    print("서브폴더목록:", sub_dir)
    
    img_folder = "yolo_data/" + dir_name + "/images"
    label_folder = "yolo_data/" + dir_name + "/labels"
   
    os.makedirs("yolo_data/" + dir_name + "/images", exist_ok=True)
    os.makedirs("yolo_data/" + dir_name + "/labels", exist_ok=True)
    
    for sub_dir_name in sub_dir:
        print("서브폴더이름:",sub_dir_name)
        sub_dir_path = os.path.join(file_path , sub_dir_name)
        print("서브폴더경로:",sub_dir_path)    
        
        img_list = glob.glob(sub_dir_path + "/*.png")
        xml_list = glob.glob(sub_dir_path + "/*.xml")
        
        for img_path in img_list:
            src_img = img_path
            dest_img = img_folder + "/" + sub_dir_name + "_" + os.path.basename(img_path)
            
            print(src_img, dest_img)
#             shutil.copy(src_img, "./" + dest_img) 
            
        for xml_path in xml_list:
            src_xml = xml_path
            dest_xml = label_folder + "/" + sub_dir_name + "_" + os.path.basename(xml_path)            
#             print(src_xml, dest_xml)
#             shutil.copy(src_xml, "./" + dest_xml)
            xml_2_yolo_txt(src_xml, dest_xml, sub_dir_name)
    
    print("파일복사 완료")          
```
